### Create Logs using your own class

```python
from datetime import datetime
import random
```

```python
help(datetime)
```

```python
print(datetime.now().date())
print(datetime.now().strftime("%H:%M:%S"))
```

```python
class App_Logger:
    def __init__(self):
        pass

    def log(self, file_object, log_message):
        self.now = datetime.now()
        self.date = self.now.date()
        self.current_time = self.now.strftime("%H:%M:%S")
        file_object.write(str(self.date) + "/" + str(self.current_time) + "\t\t" + log_message 
                          +"\n")
```

```python
logger = App_Logger()
```

```python
with open("log1.txt",'a') as fhand:
    msg = str(int((random.random())*10))+" Log message"
    logger.log(fhand,msg)
```

### Create Log using logging library

```python
import logging
```

```python
logging.basicConfig(filename = "srinivaas.txt",filemode='a',
                    format='%(asctime)s:%(levelname)s:%(message)s',datefmt="%Y-%M-%d %H:%M:%S")
for i in range(15):
    if(i%2==0):
        logging.warning("Log - Even")
    elif(i%3==0):
        logging.critical("Log - divisible by 3")
    else:
        logging.error("Log - Odd")
```
